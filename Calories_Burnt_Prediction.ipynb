## Importing Libraries and Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn import metrics
from sklearn.svm import SVC
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
df1 = pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/20240725130600/exercise.csv')
df2 = pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/20240725130529/calories.csv')

df = pd.merge(df1, df2, on="User_ID")
df.head()
```

```python
df.replace({'male': 0, 'female': 1},
           inplace=True)
df.head()
```

```python
plt.figure(figsize=(8, 8))
sb.heatmap(df.corr() > 0.9,
           annot=True,
           cbar=False)
plt.show()
```

```python
to_remove = ['Weight', 'Duration']
df.drop(to_remove, axis=1, inplace=True)
```

```python
features = df.drop(['User_ID', 'Calories'], axis=1)
target = df['Calories'].values

X_train, X_test, y_train, y_test = train_test_split(features, target,
                                      test_size=0.1,
                                      random_state=22)
X_train.shape, X_test.shape
```

```python
# Normalizing the features for stable and fast training.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Now lets train some state-of-the-art machine learning models and compare them which fit better with our data.

```python
from sklearn.metrics import mean_absolute_error as mae
models = [LinearRegression(), XGBRegressor(),
          Lasso(), RandomForestRegressor(), Ridge()]

for i in range(5):
    models[i].fit(X_train, y_train)

    print(f'{models[i]} : ')

    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(y_train, train_preds))
    print('Training Score:', r2_score(y_train, train_preds))

    val_preds = models[i].predict(X_test)
    print('Validation Error : ', mae(y_test, val_preds))
    print('Validation Score:', r2_score(y_test, val_preds))

    print()
```
